# Análise e Tratamento de Dados do Inquérito sobre WC Neutras

```python
# imports

import pandas as pd

df = pd.read_excel("WCNeutras_respostas_full.xlsx")
df[:3]
```

**Nota: Os índices da dataframe não são iguais aos do documento porque o python começa no 0 e ignora o cabeçalho, enquanto que o ficheiro começa em 1 e inclui o cabeçalho, levando a um desfazamento de 2.**

Para identificar respostas duplicadas, é necessário ignorar a 'Submission Date', pois as linhas duplicadas têm diferentes datas de submissão.

```python
df_wout_sdate = df.iloc[:,1:]
df_wout_sdate[:3]
```

Vamos então passar à identificação de linhas duplicadas acrescentando uma coluna que diz se a linha é duplicada ou não. De seguida, criamos uma dataframe apenas com as linhas duplicadas e extraímos os seus índices.

```python
df_wout_sdate["is_duplicate"]= df_wout_sdate.duplicated()

df_dup = df_wout_sdate[df_wout_sdate['is_duplicate'].values]
df_dup[:3]
```

```python
index_dup_list = df_dup.index.tolist()
index_dup_list
```

```python
# é necessário incrementar 2 a cada índice para ficar de acordo com os índices do ficheiro original
[x+2 for x in index_dup_list]
```
